```python
#import pandas as pd
from sklearn import datasets
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
x,y = datasets.load_breast_cancer(return_X_y=True)
train_x, val_x, train_y, val_y = train_test_split(x,y)
```

```python
def calcError(predictions, val_y):
    tot = 0
    for i in range(0,predictions.size-1):
        if (predictions[i] != 0 and predictions[i] != 1):
            continue
        tot+=abs(predictions[i]-val_y[i])
        if (predictions[i] != val_y[i]):
            print("HERE")
            print(predictions[i], ",",val_y[i])
        #print(predictions[i], ",",val_y[i])
    print(tot, predictions.size)
    print(1-(tot/predictions.size))
```

```python
#decisionTree
model = DecisionTreeRegressor()
model.fit(train_x,train_y)
predictions = model.predict(val_x)
calcError(predictions,val_y)
#randomForest
model = RandomForestRegressor()
model.fit(train_x,train_y)
predictions = model.predict(val_x)
calcError(predictions,val_y)
#next
```
